```python
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
import numpy as np
import statsmodels.api as sm
```

```python
dataset = pandas.read_csv('/Users/ericson/Desktop/브라이틱스/qualityData.csv')
```

```python
dataset = dataset.dropna()
```

```python
X = dataset.iloc[:, list(range(1,86))]
y = dataset.iloc[:, 86]
df = pd.DataFrame(X,y)
```

```python
df = dataset.drop('Time_Sequence', axis=1)
```

```python
df.head()
```

```python
df = df.reset_index(drop = True)
```

```python
df.head()
```

```python
N = len(df)
ratio = 0.7
np.random.seed(0)
idx_train = np.random.choice(np.arange(N), np.int(ratio * N))
idx_test = list(set(np.arange(N)).difference(idx_train))

df_train = df.iloc[idx_train]
df_test = df.iloc[idx_test]
```

```python
len(set(np.arange(N))-set(idx_train))
```

```python
len(set(idx_train))
```

```python
len(df)
```

```python
model = sm.OLS.from_formula("Y ~ " + "+".join(X.columns), data=df_train)
result = model.fit()
print(result.summary())
```

```python
pred = result.predict(df_test)

rss = ((df_test.Y - pred) ** 2).sum()
tss = ((df_test.Y - df_test.Y.mean())** 2).sum()
rsquared = 1 - rss / tss
rsquared
```

```python
pred_df = pd.concat([pred, df_test.Y], axis = 1)
```

```python
pred_df.rename(columns = {0 : 'pred' , 'Y' : 'y'})
```
